### Predictive Modelling

```python
## Target variable is_ev_dominant (Binary Classification)
## Logistic Regression, Random Forest Problem
```

```python
features=[
    'fuel_price_usd_per_liter',
    'electricity_price_usd_per_kwh',
    'charging_stations',
    'ev_subsidy_usd',
    'gdp_per_capita',
    'urban_population_percent',
    'emission_regulation_score',]
```

```python
X=df[features]  ## Intependant Variables
y=df['is_ev_dominant']  ## Target Variable
```

```python
from sklearn.model_selection import train_test_split
```

```python
## Temporal train/test split: train on years <= 2022, test on 2023+
## (avoids data leakage from future years into past predictions)
train_mask = df['year'] <= 2022
test_mask  = df['year'] > 2022

X_train = df.loc[train_mask, features]
X_test  = df.loc[test_mask,  features]
y_train = df.loc[train_mask, 'is_ev_dominant']
y_test  = df.loc[test_mask,  'is_ev_dominant']

print(f'Train size: {len(X_train)}, Test size: {len(X_test)}')
print(f'Train years: {df.loc[train_mask,"year"].min()}-{df.loc[train_mask,"year"].max()}')
print(f'Test years:  {df.loc[test_mask,"year"].min()}-{df.loc[test_mask,"year"].max()}')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.preprocessing import StandardScaler

## StandardScaler: required for Logistic Regression since features have very different scales
## (e.g. gdp_per_capita is 0-100k, electricity_price is 0-0.4)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

## class_weight='balanced': handles ~2% positive class imbalance
model = LogisticRegression(max_iter=1000, class_weight='balanced')
model.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,roc_auc_score
```

```python
pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import f1_score

## F1 score is the correct primary metric for imbalanced classification
## Accuracy is misleading when one class is ~98% of the data
print("F1 Score (macro):", f1_score(y_test, pred, average='macro'))
print("F1 Score (class 1 - EV Dominant):", f1_score(y_test, pred, pos_label=1))
print("Accuracy:", accuracy_score(y_test, pred) * 100)
```

```python
print("Classification Report:\n",classification_report(y_test,pred))
```

```python
print("ROC-AUC:", roc_auc_score(y_test, pred))
print(confusion_matrix(y_test, pred))
```

```python
importance=pd.DataFrame({'Feature':features,'Coefficient':model.coef_[0]}).sort_values(by='Coefficient',ascending=False)
print(importance)

##Interpretation:
##  Positive coefficient → increases EV dominance probability
##  Negative coefficient → decreases probability
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
## class_weight='balanced': handles class imbalance in Random Forest
rf_model = RandomForestClassifier(n_estimators=200, random_state=42, max_depth=None, class_weight='balanced')
rf_model.fit(X_train, y_train)
```

```python
rf_pred=rf_model.predict(X_test)
```

```python
print("Classification Report:\n", classification_report(y_test, rf_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, rf_pred))
print("ROC-AUC:", roc_auc_score(y_test, rf_pred))
print("F1 Score (macro):", f1_score(y_test, rf_pred, average='macro'))
print("F1 Score (class 1 - EV Dominant):", f1_score(y_test, rf_pred, pos_label=1))
print("Accuracy:", accuracy_score(y_test, rf_pred) * 100)
```

```python
rf_importance = pd.DataFrame({
    'Feature': features,
    'Importance': rf_model.feature_importances_
}).sort_values(by='Importance', ascending=False)

print(rf_importance)
```
